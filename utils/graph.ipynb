```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('../Data_sets/data.txt', delimiter=',')  # (19968, 2000)
data = data.reshape(9984, 2, 2000)
```

```python
plt.figure(figsize=(15, 6))
data_y = np.linspace(0, 2000, 2000)
for i in range(16):
    plt.plot(data_y, data[i+160, 1, :], label=f'{i}N', linewidth=0.3)
plt.legend()
plt.show()

```

```python
plt.figure(figsize=(20, 6))
data_y = np.linspace(0, 2000, 2000)
# 使用 Blues colormap 创建从浅到深的蓝色渐变
colors = plt.cm.Blues(np.linspace(0.5, 1, 16))
for i in range(16):
    plt.plot(data_y, data[160+i, 1, :], label=f'ch1_{i}', color=colors[i], linewidth=0.3)  # 添加 linewidth 参数
plt.legend()
```
